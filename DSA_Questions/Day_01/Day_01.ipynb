### Problem 1 — Find the Maximum Element

```python
def maxElement(arr):
    max = 0
    for i in range(len(arr)):
        if arr[i] > max:
            max = arr[i]
            
    return f"The max element is {max}"

arr = [3, 1, 9, 7]
print(maxElement(arr))
```

### Problem 2 — Find the Minimum Element

```python
def minElement(arr):
    min = arr[0]
    for i in range(1, len(arr)):
        if min > arr[i]:
            min = arr[i]
        else:
            continue
        
    return f"The min element is {min}"


arr = [3, 1, 9, 7]
print(minElement(arr))
```

### Problem 3 — Reverse the Array (Two-Pointer Method)

```python
def reverseArray(arr):
    left = 0
    right = len(arr)-1
    
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        
        left += 1
        right -= 1
        
    return arr

arr = [1,2,3,4,5]
print("Before", arr)
print("After ", reverseArray(arr))
```

### Problem 4 — Check if Array is Sorted

```python
def checkSort(arr):
    for i in range(len(arr)):
        if arr[i] > arr[i+1]:
            return False
        else:
            return True
        

arr = [3, 1, 9, 7]
print(checkSort(arr))
```
