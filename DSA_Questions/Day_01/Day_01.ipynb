### Problem 1 — Find the Maximum Element

```python
def maxElement(arr):
    max = 0
    for i in range(len(arr)):
        if arr[i] > max:
            max = arr[i]
            
    return f"The max element is {max}"

arr = [3, 1, 9, 7]
print(maxElement(arr))
```

### Problem 2 — Find the Minimum Element

```python
def minElement(arr):
    min = arr[0]
    for i in range(1, len(arr)):
        if min > arr[i]:
            min = arr[i]
        else:
            continue
        
    return f"The min element is {min}"


arr = [3, 1, 9, 7]
print(minElement(arr))
```

### Problem 3 — Reverse the Array (Two-Pointer Method)

```python
def reverseArray(arr):
    left = 0
    right = len(arr)-1
    
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        
        left += 1
        right -= 1
        
    return arr

arr = [1,2,3,4,5]
print("Before", arr)
print("After ", reverseArray(arr))
```

### Problem 4 — Check if Array is Sorted

```python
def checkSort(arr):
    for i in range(len(arr)):
        if arr[i] > arr[i+1]:
            return False
        else:
            return True
        

arr = [3, 1, 9, 7]
print(checkSort(arr))
```

### Problem 5 - Pair in array

```python
def arrPair(arr):
    for i in range(0, len(arr)-1):
        for j in range(i+1, len(arr)):
            print((arr[i], arr[j]), end=" ")
        print()
        
arrPair([2,4,6,8,10])
```

### 6. Count frequency of each element

```python
def count_freq_sorted(arr):
    count = 1
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            count += 1
        else:
            print(arr[i-1], "->", count)
            count = 1
    print(arr[-1], "->", count)
    
    
count_freq_sorted([1,1,2,2,2,3,4,4])
```

### 7. Find second largest element

```python
def secondLarge(arr):
    sorted_arr = sorted(arr)
    print(sorted_arr[-2])
    

secondLarge([4,9,3,8,1,2])
    
```

### 8. Remove duplicates

```python
def removeDuplicate(arr):
    distinct = []
    l = 0
    for r in range(0, len(arr)):
        if arr[r] != arr[r-1]:
            distinct.append(arr[r])
            l +=1
            
    return distinct

print(removeDuplicate([1,1,2,2,2,3,4,4]))
```

### 11 - Print Subarray

```python
def subArray(arr):
    for i in range(0, len(arr)+1):
        for j in range(i+1, len(arr)+1):
            for k in range(i, j):
                print(arr[k], end=" ")
                
            print()
        print()
            
subArray([2,4,6,8,10])
```
