```python
def longestEqual01(arr):
    zero, one = 0, 0
    res = 0
    
    diff_idx = {}  # count[1] - count[0] -> idx
    
    for i, n in enumerate(arr):
        if n == 0:
            zero += 1
        else:
            one += 1
        
        diff = one - zero
        
        # Case 1: from start to current index
        if diff == 0:
            res = max(res, one+zero)
        
        # Case 2: subarray between two same diffs
        if diff in diff_idx:
            res = max(res, i-diff_idx[diff])
        else:
            diff_idx[diff] = i
            
    return res
    
    
print(longestEqual01([0,0,1,0,0,0,1,1]))
```

```python
def sortColors(nums):
    left = 0
    mid = 0
    right = len(nums)-1
    
    while mid <= right:
        
        #  Case 1: when mid is 0
        if nums[mid] == 0:
            nums[left], nums[mid] = nums[mid], nums[left]
            mid += 1
            left += 1
            
        #  Case 1: when mid is 1
        elif nums[mid] == 1:
            mid += 1
        
        #  Case 1: when mid is 2    
        else:
            nums[mid], nums[right] = nums[right], nums[mid]
            right -= 1
            
    return nums

print(sortColors([1,0,2]))
```

```python
def searchMatrix(matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        # for row in matrix:
        #     if target in row:
        #         return True
        #     else:
        #         continue
        # return False
        
        for row in matrix:
            
            left = 0
            right = len(row)
            while left <= right:
                mid = (left+right)//2

                if row[mid] == target:
                    return True
                elif row[mid] > target:
                    right = mid -1

                else:
                    left = mid + 1

                return False
        return False
        
print(searchMatrix([[1,3]], 3))
```
