```python
def findMissingAndRepeatedValues(grid):
    seen = set()
    ans = []
    repeated = None
    missing = None
    
    for ch in grid:
        for ele in ch:
            if ele in seen:
                repeated = ele
                
            else:
                seen.add(ele)
                
    ans.append(repeated)
    
    for num in range(1, (len(grid)*len(grid))+1):
        if num not in seen:
            missing = num
            
    ans.append(missing)
    
    return ans
            
print(findMissingAndRepeatedValues([[1,3],[2,2]]))
print(findMissingAndRepeatedValues([[9,1,7],[8,9,2],[3,4,6]]))
```

```python
def merge(nums1, m, nums2, n):
    for i in range(m):
        for j in range(n):
            if nums1[i]==0:
                nums1[i] = nums2[j]
                
    return nums1

print(merge(nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3))
```
