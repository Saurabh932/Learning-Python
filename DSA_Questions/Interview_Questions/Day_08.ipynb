```python
def findMissingAndRepeatedValues(grid):
    seen = set()
    ans = []
    repeated = None
    missing = None
    
    for ch in grid:
        for ele in ch:
            if ele in seen:
                repeated = ele
                
            else:
                seen.add(ele)
                
    ans.append(repeated)
    
    for num in range(1, (len(grid)*len(grid))+1):
        if num not in seen:
            missing = num
            
    ans.append(missing)
    
    return ans
            
print(findMissingAndRepeatedValues([[1,3],[2,2]]))
print(findMissingAndRepeatedValues([[9,1,7],[8,9,2],[3,4,6]]))
```

```python
def merge(nums1, m, nums2, n):
    for i in range(m):
        for j in range(n):
            if nums1[i]==0:
                nums1[i] = nums2[j]
                
    return nums1

print(merge(nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3))
```

```python
def backend(arr):
    role = []
    for i in arr:
        if i["role"] == "Backend":
            role.append(i["name"])
            
    return role
        

employees = [
    {"id": 1, "name": "Saurabh", "role": "Backend"},
    {"id": 2, "name": "Amit", "role": "Support"},
    {"id": 3, "name": "Neha", "role": "Backend"}
]

print(backend(employees))
```

```python
def get_employee_name(arr):
    python_name = []
    
    for emp in arr:
        if "Python" in emp["skills"]:
            python_name.append(emp["name"])
            
    return python_name
    
employees = [
    {"id": 1, "name": "Saurabh", "skills": ["Python", "FastAPI"]},
    {"id": 2, "name": "Amit", "skills": ["Linux", "Support"]},
    {"id": 3, "name": "Neha", "skills": ["Python", "Django"]}
]

print(get_employee_name(employees))
```

```python
def get_employee(arr):
    employee = []
    
    for emp in arr:
        if emp["age"] > 24:
            employee.append(emp["name"])
            
    return employee

employees = [
    {"id": 1, "name": "Saurabh", "age": 25},
    {"id": 2, "name": "Amit", "age": 30},
    {"id": 3, "name": "Neha", "age": 22}
]
print(get_employee(employees))
```

```python
def get_highest_salary(arr):
    employee_name = ""
    highest_salary = 0
    
    for emp in arr:
        if highest_salary < emp['salary']:
            employee_name = emp['name']
            highest_salary = emp['salary']
        
    return employee_name, highest_salary

employees = [
    {"id": 1, "name": "Saurabh", "salary": 50000},
    {"id": 2, "name": "Amit", "salary": 40000},
    {"id": 3, "name": "Neha", "salary": 60000}
]
print(get_highest_salary(employees))
```

```python
def get_IT_employee(arr):
    employee_IT = []
    
    for emp in arr:
        if emp["department"] == "IT":
            employee_IT.append(emp["name"])

    return employee_IT
    
employees = [
    {"id": 1, "name": "Saurabh", "department": "IT"},
    {"id": 2, "name": "Amit", "department": "HR"},
    {"id": 3, "name": "Neha", "department": "IT"}
]
print(get_IT_employee(employees))
```

```python
def get_count(s):
    freq = {}
    
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
        
    return freq

print(get_count("banana"))
```
