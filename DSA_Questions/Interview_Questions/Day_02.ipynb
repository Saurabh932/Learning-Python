### Problem: Move All Zeros to the End
Asked in:

Almost every fresher interview

Very common array manipulation test

ðŸ§¾ Problem Statement

Given an array, move all 0s to the end while maintaining the relative order of non-zero elements.

Input:  [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]



```python
def moveZeros(arr):
    slow = 0
    
    for fast in range(len(arr)):
        if arr[fast] !=  0:
            arr[fast], arr[slow] = arr[slow], arr[fast]
            slow += 1
        
    return arr

print(moveZeros([2,7,0,3,0,0,0,1,5,3,0,1,0]))
```

### Problem: Count Vowels in a String

Input:

"saurabh"
Output: 3   (a, u, a)


```python
def countVowels(s):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    
    for i in range(len(s)):
        if s[i].lower() in vowels:
            count += 1
            
    return count

print(countVowels("sAurabh"))
```

### First Non-Repeating Character in a String

Input:

"swiss"


Output:

'w'


```python
def firstNonRepeatingChar(s):
    freq = {}
    
    for ch in s.lower():
        freq[ch] = freq.get(ch, 0) + 1
        
    for ch in s.lower():
        if freq[ch] == 1:
            return ch
    return None

print(firstNonRepeatingChar("saurabh"))
```

### Problem: Find the Missing Number

Given an array of size n containing numbers from 1 to n+1,
one number is missing. Find it.

Example:

Input:  [1,2,4,5]
Output: 3


```python
def missingNumber(arr):
    low, high = min(arr), max(arr)
    
    expected_sum = (low + high) * (high - low + 1) // 2
    actual_sum = sum(arr)
    
    return expected_sum - actual_sum

print(missingNumber([6,7,9,10]))
```

### Sliding Window (Revision): Maximum Sum Subarray of Size K

Example:

arr = [2, 1, 5, 1, 3, 2]
k = 3
Output = 9   ([5,1,3])


```python
def maxSumSubarray(arr, k):
    left = 0
    sum_window = []

    for right in range(1, len(arr)):
        if right - left + 1 == k:
            sum_window.append(sum(arr[left:right+1]))
            left += 1
            
    return max(sum_window)
             
print(maxSumSubarray([1,2,3,4,5], 3))
```
