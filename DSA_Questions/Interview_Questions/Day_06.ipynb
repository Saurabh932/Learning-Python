```python
def isValid(s):
    stack = []
    closeToOpen = {")":"(", "}":"{", "]":"["}
    
    for c in s:
        if c in closeToOpen:
            if stack and stack[-1] == closeToOpen[c]:
                stack.pop()
            else:
                return False
            
        else:
            stack.append(c)
            
    return True if not stack else False

print(isValid("()"))        # True
print(isValid("()[]{}"))    # True
print(isValid("(]"))        # False
print(isValid("([)]"))      # False
print(isValid("{[]}"))      # True
```

### Problem: Check if a String Has Balanced Parentheses (Only '(' and ')')

This is a simplified stack problem, often asked before the full version.

Examples
Input:  "()()"
Output: True

Input:  "(()"
Output: False

Input:  ")("
Output: False

Input:  "(())()"
Output: True


```python
def isBalanced(s):
    stack = []
    
    for ch in s:
        if ch in ["(", "{", "["]:
            stack.append(ch)
            
        else:
            if not stack:
                return False
            
            curr_ch = stack.pop()
            if curr_ch == "(":
                if ch != ")":
                    return False
                
            if curr_ch == "{":
                if ch != "}":
                    return False
                
            if curr_ch == "[":
                if ch != "]":
                    return False
                
    if stack:
        return False
    return True

print(isBalanced("()()"))     # True
print(isBalanced("(()"))      # False
print(isBalanced(")("))       # False
print(isBalanced("(())()"))   # True
```

### Problem: Next Greater Element (Basic Version)

```python
def nextGreaterElement(arr):
    n = len(arr)
    result = [-1] * n
    stack = []

    for i in range(n):
        while stack and arr[i] > arr[stack[-1]]:
            idx = stack.pop()
            result[idx] = arr[i]
        stack.append(i)

    return result

                
print(nextGreaterElement([4, 5, 2, 10]))
print(nextGreaterElement([3,2,1]))
```

### Problem: Stock Buy and Sell (Max Profit â€“ Single Transaction)
Problem

Given prices of a stock on different days, find the maximum profit you can achieve by buying once and selling once.

Examples
Input:  [7,1,5,3,6,4]
Output: 5   (buy at 1, sell at 6)

Input:  [7,6,4,3,1]
Output: 0


```python
def maxProfit(arr):
    min_price = float('inf')
    max_profit = 0
    
    for price in arr:
        if price < min_price:
            min_price = price
        else:
            max_profit = max(max_profit, price-min_price)
            
    return max_profit

print(maxProfit([7,1,5,3,6,4]))  # 5
print(maxProfit([7,6,4,3,1]))    # 0
print(maxProfit([1]))            # 0
print(maxProfit([2,4,1]))        # 2
```
