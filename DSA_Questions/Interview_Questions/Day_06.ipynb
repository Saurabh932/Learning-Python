```python
def isValid(s):
    stack = []
    closeToOpen = {")":"(", "}":"{", "]":"["}
    
    for c in s:
        if c in closeToOpen:
            if stack and stack[-1] == closeToOpen[c]:
                stack.pop()
            else:
                return False
            
        else:
            stack.append(c)
            
    return True if not stack else False

print(isValid("()"))        # True
print(isValid("()[]{}"))    # True
print(isValid("(]"))        # False
print(isValid("([)]"))      # False
print(isValid("{[]}"))      # True
```

### Problem: Check if a String Has Balanced Parentheses (Only '(' and ')')

This is a simplified stack problem, often asked before the full version.

Examples
Input:  "()()"
Output: True

Input:  "(()"
Output: False

Input:  ")("
Output: False

Input:  "(())()"
Output: True


```python
def isBalanced(s):
    stack = []
    
    for ch in s:
        if ch in ["(", "{", "["]:
            stack.append(ch)
            
        else:
            if not stack:
                return False
            
            curr_ch = stack.pop()
            if curr_ch == "(":
                if ch != ")":
                    return False
                
            if curr_ch == "{":
                if ch != "}":
                    return False
                
            if curr_ch == "[":
                if ch != "]":
                    return False
                
    if stack:
        return False
    return True

print(isBalanced("()()"))     # True
print(isBalanced("(()"))      # False
print(isBalanced(")("))       # False
print(isBalanced("(())()"))   # True
```

```python
def nextGreaterElement(arr):
    greater_ele = []
    
    left = 0
    right = 1
    
    while left < right and right < len(arr):
        if arr[right] > arr[left]:
            greater_ele.append(arr[right])
            left+=1
            right+=1
            
        else:
            right += 1
            
    return greater_ele
                
print(nextGreaterElement([4, 5, 2, 10]))
```
