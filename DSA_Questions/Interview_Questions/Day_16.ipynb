```python
def merge(intervals):
    # sorting the interval based on the first element of the intervcal
    intervals.sort(key= lambda interval:interval[0])
    
    merged = []
    
    for interval in intervals:
        # check overlap if mergerd is not empty or last element of last merged-intrval is less than first element of interval
        if not merged or merged[-1][1] < interval[0]:
            merged.append(interval)
            
        else:
            merged[-1] = [merged[-1][0], max(merged[-1][1], interval[1])] 
    
    
    return merged

print(merge([[2,6],[1,3],[8,10],[15,18]]))
```

```python
def lengthOfLongestSubstring(s):
    seen = set()
    slow = 0
    max_len = 0
    
    for fast in range(len(s)):
        while s[fast] in seen:
            seen.remove(s[slow])
            slow += 1
            
        seen.add(s[fast])
        max_len = max(max_len, (fast-slow + 1))
            
    return max_len

print(lengthOfLongestSubstring("abcabcbb"))
```
