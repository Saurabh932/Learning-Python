### Find the Missing Number (0 to n)

You are given an array containing n distinct numbers taken from the range 0 to n.
Exactly one number is missing. Find it.

Examples
Input:  [3,0,1]
Output: 2

Input:  [0,1]
Output: 2



```python
def missingNumber(arr):
    actual = len(arr) * (len(arr)+1) // 2
    org_arr = sum(arr)
    
    return actual - org_arr

print(missingNumber([9,6,4,2,3,5,7,0,1]))
```

### Problem: Check if Two Strings Are Rotations of Each Other
ðŸ§¾ Problem Statement

Given two strings, check if one string is a rotation of the other.

Examples
Input:  s1 = "abcd", s2 = "cdab"
Output: True

Input:  s1 = "abcd", s2 = "acbd"
Output: False


```python
def rotatedString(s1, s2):
    if len(s1) != len(s2):
        return False
    
    rot_str = s1+s1
    
    if s2 in rot_str:
         return True
    return False
 
 
print(rotatedString("a", "a"))
```

### Problem: Find the Second Smallest Element in an Array

This is a variant of a question you already did.

Examples
Input:  [12, 13, 1, 10, 34, 1]
Output: 10

Input:  [5, 5, 5]
Output: -1


```python
def secondSmallest(arr):
    min_ele = float("inf")
    sec_min = float("inf")
    
    for i in range(len(arr)):
        if arr[i] < min_ele:
            sec_min = min_ele
            min_ele = arr[i]
            
        elif arr[i] != min_ele and arr[i] < sec_min:
            sec_min = arr[i]
            
    return sec_min if sec_min != float("inf") else -1

print(secondSmallest([12,13,1,10,34,1]))  # 10
print(secondSmallest([5,5,5]))            # -1
```

### Problem: Reverse Each Word in a String

This is different from reversing word order.

Examples
Input:  "I love coding"
Output: "I evol gnidoc"

Input:  "hello world"
Output: "olleh dlrow"


```python
def revEachWord(s):
    s_arr = []
    for i in s.split():
        s_arr.append(i[::-1])
        
    return " ".join(s_arr)

print(revEachWord("hello World"))    
```

### Problem: Find the Maximum Number of Consecutive 1s
Examples
Input:  [1,1,0,1,1,1]
Output: 3

Input:  [1,0,1,1,0,1]
Output: 2

Input:  [0,0,0]
Output: 0


```python
def maxConsecutiveOnes(arr):
    max_count = 0
    current = 0
    
    for num in arr:
        if num==1:
            current += 1
            max_count = max(current, max_count)
            
        else:
            current = 0
            
    return max_count

print(maxConsecutiveOnes([1,1,0,1,1,1]))  # 3
print(maxConsecutiveOnes([1,0,1,1,0,1]))  # 2
print(maxConsecutiveOnes([0,0,0]))        # 0
```
