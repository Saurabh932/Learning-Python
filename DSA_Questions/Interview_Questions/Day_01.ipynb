### Problem: Find the Second Largest Element in an Array

Problem Statement

Given an array of integers, find the second largest element.

Input:  [12, 35, 1, 10, 34, 1]
Output: 34


Another example:

Input:  [10, 10, 10]
Output: -1   (or "not possible")


```python
def secondLargest(arr):
    # sort_arr = sorted(arr, reverse=True)
    # return sort_arr[1]

    max_ele = float("-inf")
    s_max = float("-inf")
    for i in range(len(arr)):
        if arr[i] > max_ele:
            s_max = max_ele
            max_ele = arr[i]
            
        elif arr[i] != max_ele and arr[i] > s_max:
            s_max = arr[i]
        
    return s_max if s_max != float("-inf") else None
            

print(secondLargest([5, 5, 5]) )
```

### Problem: Rotate an Array by K Positions (Right Rotation)

Example:

Input:  arr = [1,2,3,4,5]
K = 2
Output: [4,5,1,2,3]


```python
def rotateArray(arr, k):
    while k!=0:
        last_ele = arr.pop()
        arr.insert(0, last_ele)
        k -= 1
        
    return arr

print(rotateArray([1,2,3,4,5], 2))
```

### Reverse Words in a String

Input:

"I love coding"


Output:

"coding love I"


```python
def reverseWords(s):
    # words = s.split()
    # words.reverse()
    # return " ".join(words)
        
    s = s[::-1]              # reverse entire string
    words = []

    for word in s.split():
        words.append(word[::-1])

    return " ".join(words)

print(reverseWords("I love coding"))
```

### Check if Two Strings Are Anagrams

Input:

"listen", "silent" → True
"hello", "bello"  → False


```python
def isAnagram(s1, s2):
    if len(s1) != len(s2):
        return False
    
    freq = {}
    for ch in s1:
        freq[ch] = freq.get(ch, 0) + 1
        
    for ch in s2:
        if ch not in freq or freq[ch] == 0:
            return False
        freq[ch] -= 1
        
    return True
        

print(isAnagram("bello", "hello"))
```

```python
def isSorted(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            return False
        
    return True

print(isSorted([6,1,9,2,3,4,5]))
```
