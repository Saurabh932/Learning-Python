```python
def buildPrefix(arr):
    prefix = [0]*len(arr)
    prefix[0] = arr[0]
    
    for i in range(1, len(arr)):
        prefix[i] = prefix[i-1] + arr[i]
        
    return prefix

def rangeSum(prefix, left, right):
    result = prefix[right] - (prefix[left - 1] if left>0 else 0)
    
    return result

arr = [2,4,6,8,10]
prefix = buildPrefix(arr)

print(rangeSum(prefix, 0, 2))  # 12
print(rangeSum(prefix, 1, 3))  # 18
print(rangeSum(prefix, 2, 4))  # 24
```

```python
def nextPermutation(arr):
    # Finding the pivot element
    pvt = -1
    for i in range(len(arr)-2, -1, -1):
        if arr[i] < arr[i+1]:
            pvt = i
            break
        
    if pvt == -1:
        return reversed(arr)
    
    # Swapping pivot element with next largest number
    for i in range(len(arr)-1, pvt, -1):
        if arr[i] > arr[pvt]:
            arr[i], arr[pvt] = arr[pvt], arr[i]
            break
        
    # Reversing the elements after the pivot
    left = pvt+1
    right = len(arr) - 1
    
    while left <= right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
        
    return arr


print(nextPermutation([1,2,3]))
```

```python
def isPalindrome(s):
    
    left = 0
    right = len(s)-1
    
    while left < right:     
        if not s[left].isalnum():
            left += 1
            continue
            
        if not s[right].isalnum():
            right -= 1
            continue
        
        if s[right].lower() != s[left].lower():
            return False
        
        left += 1
        right -= 1
        
    return True

print(isPalindrome("A man, a plan, a canal: Panama"))
```
