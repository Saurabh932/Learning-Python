### Problem: Find the First Occurrence of an Element in a Sorted Array

Example:

Input:  arr = [1,2,2,2,3,4], target = 2
Output: 1


```python
def firstOccurance(arr, target):
    left = 0
    right  = len(arr)-1
    
    while left<=right:
        mid = (left+right)//2
        
        if arr[mid] == target and (mid == 0 or arr[mid-1] != target):
            return mid
        
        if arr[mid] >= target:
            right = mid-1
            
        else:
            left = mid+1
            
    return -1


print(firstOccurance([1,2,2,2,3,4], 2))  # 1
print(firstOccurance([1,1,1,1], 1))      # 0
print(firstOccurance([1,2,3,4], 5))      # -1
print(firstOccurance([2], 2))            # 0
print(firstOccurance([2], 1))            # -1

```

### Problem: Find the Last Occurrence of an Element in a Sorted Array

Example:

Input:  [1,2,2,2,3,4], target = 2
Output: 3


```python
def lastOccurrence(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target and (mid == len(arr) - 1 or arr[mid + 1] != target):
            return mid

        if arr[mid] <= target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


print(lastOccurrence([1,2,2,2,3,4], 2)) 
```

### Problem: Minimum Sum Subarray of Size K
ðŸ§¾ Problem Statement

Given an array of integers and an integer k, find the minimum sum of any contiguous subarray of size k.

Examples
Input:  arr = [2, 1, 5, 1, 3, 2], k = 3
Output: 7    # [1,5,1]

Input:  arr = [1, 2, 3, 4], k = 2
Output: 3    # [1,2]

Input:  arr = [10, 5, 2, 7, 8, 7], k = 3
Output: 14   # [5,2,7]


```python
def minSumSubarray(arr, k):
    # Sum of first window
    window_sum = sum(arr[:k])
    min_sum = window_sum

    left = 0
    for right in range(k, len(arr)):
        window_sum = window_sum - arr[left] + arr[right]
        left += 1
        min_sum = min(min_sum, window_sum)

    return min_sum

    
print(minSumSubarray([2,1,5,1,3,2], 3))   # 7
print(minSumSubarray([1,2,3,4], 2))       # 3
print(minSumSubarray([10,5,2,7,8,7], 3))  # 14

```
