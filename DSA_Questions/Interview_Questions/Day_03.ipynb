### Problem: Check if a String is a Palindrome

```python
def isPalindrome(s):
    # Method 1 
    # s_rev = s[::-1]
    # return True if s_rev == s else False

    # Mthod 2
    left = 0
    right = len(s)-1
    while left <= right:
        if s[left] == s[right]:
            left += 1
            right -= 1
        
        else:
            return False
    
    return True
    
print(isPalindrome("racecar"))
```

### Given a sorted array, remove duplicates in-place and return the new length.

Example:

Input:  [1,1,2,2,3]
Output: [1,2,3]


```python
def removeDuplicates(arr):
    # arr_set = set(arr)
    # return arr_set

    slow = 0
    for fast in range(1, len(arr)):
        if arr[fast] == arr[slow]:
            continue
        else:
            slow += 1
            arr[fast], arr[slow] = arr[slow], arr[fast]
            
    return arr[:slow+1]
    
print(removeDuplicates([1,1,2,2,3]))
```

### Problem: Find the First Duplicate Element

Example:

Input:  [1,2,3,2,4]
Output: 2


```python
def firstDuplicate(arr):
    freq = {}
    
    for ch in arr:
        freq[ch] = freq.get(ch, 0)+1
        
    for ch in arr:
        if freq[ch] > 1:
            return ch
        
    return -1

print(firstDuplicate([1,3,2,3,3,3,2,4]))
```

### Problem: Remove Duplicate Characters from a String

Example:

Input:  "programming"
Output: "progamin"


Rules:

Keep first occurrence

Maintain original order

Case-sensitive for now

```python
def removeDuplicateChars(s):
    """Method-1"""
    seen = set()
    result = []
    
    for ch in s.lower():
        if ch not in seen:
            seen.add(ch)
            result.append(ch)
            
    return "".join(result)

print(removeDuplicateChars("programming"))
```

### Problem: Find the Majority Element

An element is a majority if it appears more than n/2 times.

Example:

Input:  [2,2,1,2,3,2,2]
Output: 2


```python
def majorityElement(arr):
    freq = {}
    for ch in arr:
        freq[ch] = freq.get(ch, 0) + 1
        
    for ch in arr:
        if freq[ch] > len(arr)/2:
            return ch
    return -1

print(majorityElement([1]))
```
