```python
def subarrayWithSum(arr, target):
    left = 0
    curr_sum = 0
    
    for right in range(len(arr)):
        curr_sum += arr[right]
        
        while curr_sum > target and left <= right:
            curr_sum -= arr[left]
            left += 1
            
        if curr_sum == target:
            return (left, right)
        
    return (-1,-1)
        
print(subarrayWithSum([1, 4, 20, 3, 10, 5], 33))
```

```python
def maxSubarraySum(arr):
    curr_sum = 0
    max_sub = float("-inf")
    
    for i in range(len(arr)):
        curr_sum += arr[i]
        max_sub = max(curr_sum, max_sub)
        
        if curr_sum < 0:
            curr_sum = 0
            
    return max_sub

print(maxSubarraySum([-2,1,-3,4,-1,2,1,-5,4]))  # 6
print(maxSubarraySum([1]))                     # 1
print(maxSubarraySum([5,4,-1,7,8]))             # 23
print(maxSubarraySum([-1,-2,-3,-4]))            # -1
```
