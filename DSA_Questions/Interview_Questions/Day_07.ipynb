### Problem: Remove Adjacent Duplicates in a String
ðŸ§¾ Problem Statement

Given a string, repeatedly remove adjacent duplicate characters until no such duplicates exist.

```python
def removeAdjacentDuplicates(s):
    stack = []
    
    for c in s:
        if stack and stack[-1] == c:
            stack.pop()
        else:
            stack.append(c)
            
    return "".join(stack)

print(removeAdjacentDuplicates("abbaca"))  # "ca"
print(removeAdjacentDuplicates("azxxzy"))  # "ay"
print(removeAdjacentDuplicates("a"))       # "a"

            
```

### Problem: Find the Missing and Repeating Number

You are given an array of size n containing numbers from 1 to n.
One number is missing and one number is repeating.

ðŸ§ª Test Cases
Test Case 1
arr = [1,3,3]
Output: (missing=2, repeating=3)


```python
def findMissingAndRepeating(arr):
    '''Method 1'''
    set_arr = set(arr)
    n_sum = (len(arr)*(len(arr)+1))//2
    missing = n_sum - sum(set_arr)
    
    return f"repeating - {sum(arr) - sum(set_arr)}, missing - {missing}"



print(findMissingAndRepeating([4,3,6,2,1,1]))
```

### Problem: Find the Equilibrium Index
ðŸ§¾ Problem Statement

An equilibrium index is an index such that the sum of elements on the left of that index is equal to the sum of elements on the right.

The element at the index itself is not included in either sum.

If multiple exist, return the first one.

If none exist, return -1.

```python
def equilibriumIndex(arr):
    if len(arr) == 1:
        return 0
    
    
    total_sum = sum(arr)
    left_sum = 0
    
    for i in range(len(arr)):
        right_sum = total_sum - left_sum - arr[i]
        if left_sum == right_sum:
            return i
        
        left_sum += arr[i]
        
    return -1
    
print(equilibriumIndex([1,3,5,2,2]))  # 2
print(equilibriumIndex([1,2,3]))      # -1
print(equilibriumIndex([2,4,6]))      # 1
print(equilibriumIndex([1]))          # 0
```

```python
def reverse_string(s: str) -> str:
    rev_str = ""
    
    for i in range(len(s)-1, -1, -1):
        rev_str+=s[i]
        
    return rev_str
    
    
print(reverse_string("saurabh"))
```

```python
def is_palindrome(s: str) -> bool:
    left = 0
    right = len(s)-1
    
    while left < right:
        if s[left] == s[right]:
            left += 1
            right -= 1
        else:
            return False
        
    return True

print(is_palindrome("racecar"))
```

```python
def first_non_repeating(nums: list[int]) -> int | None:
    freq = {}
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
        
    for num in nums:
        if freq[num] == 1:
            return num
            
print(first_non_repeating([4, 5, 1, 2, 0, 4]))
```

```python
def second_largest(nums: list[int]) -> int:
    max_ele = float("-inf")
    sec_max_ele = float("-inf")
    
    for num in nums:
        if num > max_ele:
            sec_max_ele = max_ele
            max_ele = num
            
    return sec_max_ele if sec_max_ele != float("-inf") else None

print(second_largest([4, 5, 1, 2, 0]))
```

```python
def find_missing(nums: list[int], n: int) -> int:
    total_sum = (n*(n+1))//2
    n_sum = sum(nums)
    
    missing = total_sum - n_sum
    
    return missing

print(find_missing([1, 2, 4, 5], 5))
```

```python
def is_anagram(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False
    
    freq1 = {}
    for ch in s1:
        freq1[ch] = freq1.get(ch, 0) + 1
        
    for ch in s2:
        if ch not in freq1:
            return False
            
        freq1[ch] -= 1
    
    return all(count == 0 for count in freq1.values())

print(is_anagram("form", "morf"))
```

```python
def first_duplicate(nums: list[int]) -> int | None:
    seen = set()
    
    for num in nums:
        if num in seen:
            return num
        
        seen.add(num)
        
print(first_duplicate([1, 2, 3, 2, 4]))
```

```python
def majority_element(nums: list[int]) -> int:
    freq = {}
    
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
        
    for num in nums:
        if freq[num] > len(nums)//2:
            return num
        
print(majority_element([3, 3, 4, 2, 3, 3, 3]))
```

```python
def move_zeros(nums: list[int]) -> None:
    slow = 0
    
    for fast in range(len(nums)):
        if nums[fast] != 0:
            nums[slow], nums[fast] = nums[fast], nums[slow]
            slow += 1
            
    return nums

print(move_zeros([0, 1, 0, 3, 12]))
```

```python
def equilibrium_index(nums: list[int]) -> int:
    total_sum = sum(nums)
    left_sum = 0
    
    for num in range(len(nums)):
        right_sum = total_sum - left_sum - nums[num]
        if right_sum == left_sum:
            return num
        
        left_sum += nums[num]
        
print(equilibrium_index([1, 3, 5, 2, 2]))
```

```python
def longest_word(sentence: str) -> str:
    longest = ""
    
    for word in sentence.split():
        if len(word) > len(longest):
            longest = word
            
    return longest


print(longest_word("I love backend development"))
```

```python
def char_frequency(s: str) -> dict:
    freq = {}
    
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
        
    return freq

print(char_frequency("banana"))
```

```python
def two_sum(nums: list[int], target: int) -> tuple[int, int]:
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i]+nums[j] == target:
                return nums[i], nums[j]
        
        
print(two_sum([2, 7, 11, 15], 17))
```

```python
def remove_duplicates(nums: list[int]) -> list[int]:
    seen = set()
    result = []
    
    for num in nums:
        if num not in seen:
            seen.add(num)
            result.append(num)
            
    return result

print(remove_duplicates([1, 2, 2, 3, 1]))
```

```python
def is_prime(n: int) -> bool:
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n%i == 0:
            return False
        
    return True
    
print(is_prime(10))
```

```python
def count_vowels(s: str) -> int:
    vowel = ['a', 'e', 'i', 'o', 'u']
    count = 0
    
    for ch in s.lower():
         if ch in vowel:
             count += 1
             
    return count

print(count_vowels("Saurabh"))
```

```python
def is_power_of_two(n: int) -> bool:
    return n>0 and (n & (n-1)) == 0

print(is_power_of_two(10))
```
