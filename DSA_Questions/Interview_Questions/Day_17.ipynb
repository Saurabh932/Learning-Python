```python
def reverseWords(s):
    """
    :type s: str
    :rtype: str
    """
    s = s[::-1]
    words = []
    
    for word in s.split():
        words.append(word[::-1])
    
    return " ".join(words)

print(reverseWords("the sky is blue"))
```

```python
import math

def productExceptSelf(nums):
    """
    :type nums: List[int]
    :rtype: List[int]
    """
    ans = [1]*len(nums)
    
    # Prefix product
    prefix = 1
    for i in range(len(nums)):
        ans[i] = prefix
        prefix *= nums[i]
        
    
    suffix = 1
    for i in range(len(nums)-1, -1, -1):
        ans[i] *= suffix
        suffix *= nums[i]
        
    return ans

print(productExceptSelf([-1,1,0,-3,3]))
```
