##### Find the sum of every window of size 2:
arr = [2, 3, 4, 5]
Output:
[5, 7, 9]


```python
def sumk2(arr):
    sum_list = []
    left = 0
    for right in range(1, len(arr)):
        curr_sum = arr[left] + arr[right]
        sum_list.append(curr_sum)
        left += 1
        
    return sum_list

print(sumk2([2, 3, 4, 5]))
```

#### Count even in window

```python
def evenCount(arr, k):
    left = 0
    count = 0
    result = []
    for right in range(len(arr)):
        if arr[right]%2 == 0:
            count += 1
            
        if right - left + 1 == k:
            result.append(count)
            
            if arr[left]%2 == 0:
                count -= 1
            
            left += 1
            
    return result
            
print(evenCount([2, 1, 2, 1, 5, 5, 2], 4))
```

### Maximum Sum Subarray of Size K
üìå Problem

Given an array of integers and an integer K,
find the maximum sum of any subarray of length K.

Example:

arr = [2, 1, 5, 1, 3, 2]
K = 3


| Iter | left | right | arr[right] Added | Window Values | Window Sum | Window Size | Action              | max_sum |
| ---- | ---- | ----- | ---------------- | ------------- | ---------- | ----------- | ------------------- | ------- |
| 0    | 0    | 0     | +2               | [2]           | 2          | 1           | Expand              | 2       |
| 1    | 0    | 1     | +1               | [2,1]         | 3          | 2           | Expand              | 3       |
| 2    | 0    | 2     | +5               | [2,1,5]       | 8          | 3           | Evaluate ‚Üí Remove 2 | 8       |
| ‚Üí    | 1    | 2     | -2               | [1,5]         | 6          | 2           | Slide Left          | ‚Äî       |
| 3    | 1    | 3     | +1               | [1,5,1]       | 7          | 3           | Evaluate ‚Üí Remove 1 | 8       |
| ‚Üí    | 2    | 3     | -1               | [5,1]         | 6          | 2           | Slide Left          | ‚Äî       |
| 4    | 2    | 4     | +3               | [5,1,3]       | **9**      | 3           | Evaluate ‚Üí Remove 5 | **9**   |
| 5    | 3    | 5     | +2               | [1,3,2]       | 6          | 3           | Evaluate ‚Üí Remove 1 | 9       |


```python
def maxSumSubarray(arr, k):
    left = 0
    window_sum = 0
    max_sum = 0
    
    for right in range(len(arr)):
        window_sum += arr[right]
        
        if right - left + 1 == k:
            max_sum = max(max_sum, window_sum)
            window_sum -= arr[left]
            left += 1
            
    return max_sum

print(maxSumSubarray([2, 1, 5, 1, 3, 2], 3))
```

### Problem 2: First Negative Number in Every Window of Size K

Example:

arr = [12, -1, -7, 8, -15, 30, 16, 28]
K = 3


Windows & answers:

[12, -1, -7] ‚Üí -1  
[-1, -7, 8] ‚Üí -1  
[-7, 8, -15] ‚Üí -7  
[8, -15, 30] ‚Üí -15  
[-15, 30, 16] ‚Üí -15  
[30, 16, 28] ‚Üí 0 (no negative ‚Üí use 0)


Output:

[-1, -1, -7, -15, -15, 0]

Strategy

We maintain:

Sliding window: left & right

Queue: stores only negative numbers inside window

At each full window (size = K):

1Ô∏è‚É£ If queue has negative ‚Üí answer = queue[0]
2Ô∏è‚É£ Else ‚Üí answer = 0
3Ô∏è‚É£ If element at left = queue front ‚Üí pop from queue
4Ô∏è‚É£ Move left ‚Üí slide window

üìä Full Table Simulation (Your Favorite! üòé)
| Iter | left | right | Window        | Queue (negatives) | Window size | Output     | Action                     |
| ---- | ---- | ----- | ------------- | ----------------- | ----------- | ---------- | -------------------------- |
| 0    | 0    | 0     | [12]          | []                | 1           | ‚Äî          | Expand right               |
| 1    | 0    | 1     | [12, -1]      | [-1]              | 2           | ‚Äî          | Expand right               |
| 2    | 0    | 2     | [12, -1, -7]  | [-1, -7]          | 3           | **-1**     | Output first queue element |
| ‚Üí    | 1    | 2     | [-1, -7]      | [-1, -7]          | slide       | ‚Äî          | left++                     |
| 3    | 1    | 3     | [-1, -7, 8]   | [-1, -7]          | 3           | **-1**     | Output                     |
| ‚Üí    | 2    | 3     | [-7, 8]       | [-7]              | slide       | remove -1  | left++                     |
| 4    | 2    | 4     | [-7, 8, -15]  | [-7, -15]         | 3           | **-7**     | Output                     |
| ‚Üí    | 3    | 4     | [8, -15]      | [-15]             | slide       | remove -7  | left++                     |
| 5    | 3    | 5     | [8, -15, 30]  | [-15]             | 3           | **-15**    | Output                     |
| ‚Üí    | 4    | 5     | [-15, 30]     | [-15]             | slide       | ‚Äî          | left++                     |
| 6    | 4    | 6     | [-15, 30, 16] | [-15]             | 3           | **-15**    | Output                     |
| ‚Üí    | 5    | 6     | [30, 16]      | []                | slide       | remove -15 | left++                     |
| 7    | 5    | 7     | [30, 16, 28]  | []                | 3           | **0**      | No negatives               |
| ‚Üí    | 6    | 7     | [16, 28]      | []                | slide       | ‚Äî          | left++                     |



```python
from collections import deque

def firstNegativeWindow(arr, k):
    left = 0
    q = deque()
    result = []
    
    for right in range(len(arr)):
        if arr[right] < 0:
            q.append(arr[right])
            
        if right - left + 1 == k:
            if q:
                result.append(q[0])
            else:
                result.append(0)
                
            if arr[left]<0 and arr[left] == q[0]:
                q.popleft()
                
            left += 1
            
    return result


print(firstNegativeWindow([12, -1, -7, 8, -15, 30, 16, 28], 3))
```

### Maximum Element in Every Window of Size K

```python
from collections import deque

def maxElementWindow(arr, k):
    left = 0
    q = deque()
    result = []
    
    for right in range(len(arr)-1):
        
        while q and q[-1] < arr[right]:
            q.pop()
        q.append(arr[right])

        if right - left + 1 == k:
            result.append(q[0])

            if arr[left] == q[0]:
                q.popleft()

            left += 1
            
                
    return result

print(maxElementWindow([1, 3, -1, -3, 5, 3, 6, 7], 3))
```
