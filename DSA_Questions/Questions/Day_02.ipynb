### 1 -  Reverse an Array using Two Pointers

```python
def revArray(arr):
    left = 0
    right = len(arr)-1
    while left<right:
        arr[left], arr[right] = arr[right], arr[left]
        
        left += 1
        right -= 1
        
    return arr

print(revArray([1,2,3,4,5]))
```

### 2 - Check if an array is Palindrome

```python
def checkPalindrome(arr):
    left = 0
    right = len(arr)-1
    while left < right:
        if arr[left] == arr[right]:
            left += 1
            right -= 1
        else:
            return False
        
    return True

print(checkPalindrome([1,2,3,2,1]))
```

### 3 - Find a Pair in a SORTED Array Whose Sum Equals a Target

```python
def pairSum(arr, target):

    left = 0
    right = len(arr)-1
    while left< right:
        sum = arr[left] + arr[right]
        
        if sum == target:
            return [left, right]
        
        if sum < target:
            left += 1
        else:
            right += 1
        

    

print(pairSum([1,2,3,4,5], 9))
```

### 4: Remove duplicates from sorted array

```python
def duplicate(arr):
    slow = 0
    for fast in range(1,len(arr)):
        if arr[fast] != arr[slow]:
            slow += 1
            arr[slow] = arr[fast]
            
    
    
    return arr[:slow+1]
    

print(duplicate([1, 1, 2, 2, 2, 3, 4, 4]))
```

#### 5. Dutch National Flag Problem
Given an array containing only 0s, 1s, and 2s, sort it in one pass without using any sorting function.

```python
def dnf(arr):
    low = 0
    mid = 0
    high = len(arr)-1
    
    while mid < high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            mid += 1
            low += 1
            
        elif arr[mid] ==1:
            mid += 1
            
        else:
            arr[high], arr[mid] = arr[mid], arr[high]
            high -= 1
            
    return arr

print(dnf([0, 0, 1, 1, 2, 2])) 
```

### 6. Trapping Rain Water Problem

```python
def trapWater(arr):
    left= 0
    right = len(arr)-1
    left_max = arr[0]
    right_max = arr[len(arr)-1]
    total_water = 0
    
    while left < right:
        trapped = 0
        if arr[left] < arr[right]:
            trapped = left_max - arr[left]
            total_water += trapped
            left += 1
            trapped = 0
            
        else:
            trapped = right_max - arr[right]
            total_water += trapped
            right -= 1
            trapped = 0
            
    return total_water

print(trapWater([4,2,0,3,2,5]))
```

| Final   | left == right |       |      |      |       |            |      | **49**   | Loop ends                                         |


```python
def maxWater(arr):
    left = 0
    right = len(arr)-1
    area = 0
    while left < right:
        area = max((min(arr[left], arr[right]) * (right-left)), area)
        
        if arr[left] < arr[right]:
            left += 1
        else:
            right -= 1
            
    return  area


print(maxWater([1,8,6,2,5,4,8,3,7]))
```

### 8. 3-SUM
Given an array nums, return all unique triplets [a, b, c] such that:

a + b + c == 0


Example:

Input: [-1, 0, 1, 2, -1, -4]
Output:
[
  [-1, -1, 1],
  [-1, 0, 1]
]

```python
def sum3(arr):
    arr = sorted(arr)
    unique = []
    
    for i in range(len(arr)-1):
        left = i+1
        right = len(arr)-1
        
        while left < right:
            curr_sum = arr[i] + arr[left] + arr[right]
            triplet = [arr[i], arr[left], arr[right]]
            
            if curr_sum == 0 and triplet not in  unique:
                unique.append(triplet)
                left += 1
                right -= 1
            
            if curr_sum < 0:
                left += 1
                
            else:
                right -= 1
    
    return unique
    
    
print(sum3([-1, 0, 1, 2, -1, -4]))
```

### 9: Sort Squares of a Sorted Array


| Iter | left         | right | arr[left] | arr[right] | left² | right² | Larger | result[position]     | position→ | Move Pointer |
| ---- | ------------ | ----- | --------- | ---------- | ----- | ------ | ------ | -------------------- | --------- | ------------ |
| Init | 0            | 4     | -4        | 3          | 16    | 9      | 16     | result[4] = 16       | 4→3       | left++       |
| 1    | 1            | 4     | -2        | 3          | 4     | 9      | 9      | result[3] = 9        | 3→2       | right--      |
| 2    | 1            | 3     | -2        | 1          | 4     | 1      | 4      | result[2] = 4        | 2→1       | left++       |
| 3    | 2            | 3     | 0         | 1          | 0     | 1      | 1      | result[1] = 1        | 1→0       | right--      |
| 4    | 2            | 2     | 0         | 0          | 0     | 0      | 0      | result[0] = 0        | 0→(-1)    | left++       |
| End  | left > right | —     | —         | —          | —     | —      | —      | Final → [0,1,4,9,16] | —         | Loop ends    |


```python
def squareSum(arr):
    left = 0
    right = len(arr)-1
    result = [0] * len(arr)
    pos = right
    
    while left <= right:
        left_sq = arr[left] ** 2
        right_sq = arr[right] ** 2
        
        if left_sq > right_sq:
            result[pos] = left_sq
            left += 1
            
        else:
            result[pos] = right_sq
            right -= 1
            
        pos -= 1
               
    return result

print(squareSum([-4, -2, 0, 1, 3]))
```

### 10:rge Two Sorted Arrays

Given two sorted arrays, merge them into one sorted array.

| Iter | i                        | j | A[i] | B[j] | Smaller     | Added to result | Pointer Moved | **Array After Iteration** |
| ---- | ------------------------ | - | ---- | ---- | ----------- | --------------- | ------------- | ------------------------- |
| Init | 0                        | 0 | 1    | 2    | 1           | 1               | i++           | `[1]`                     |
| 1    | 1                        | 0 | 3    | 2    | 2           | 2               | j++           | `[1, 2]`                  |
| 2    | 1                        | 1 | 3    | 4    | 3           | 3               | i++           | `[1, 2, 3]`               |
| 3    | 2                        | 1 | 5    | 4    | 4           | 4               | j++           | `[1, 2, 3, 4]`            |
| 4    | 2                        | 2 | 5    | 6    | 5           | 5               | i++           | `[1, 2, 3, 4, 5]`         |
| 5    | 3                        | 2 | 7    | 6    | 6           | 6               | j++           | `[1, 2, 3, 4, 5, 6]`      |
| 6    | 3                        | 3 | 7    | —    | A leftovers | 7               | i++           | `[1, 2, 3, 4, 5, 6, 7]`   |
| End  | i > len(A) or j > len(B) | — | —    | —    | —           | —               | Loop End      | Final: `[1,2,3,4,5,6,7]`  |


```python
def mergeArray(arr1, arr2):
    i = 0
    j = 0
    result = []
    
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1
            
    while i < len(arr1):
        result.append(arr1[i])
        i += 1
        
    while j < len(arr2):
        result.append(arr2[j])
        j += 1
            
    return result

print(mergeArray([1,3,5,7], [2,4,6]))
```
