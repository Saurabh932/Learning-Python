### Problem 1 — Find the Maximum Element

```python
def maxElement(arr):
    max = 0
    for i in range(len(arr)):
        if arr[i] > max:
            max = arr[i]
            
    return f"The max element is {max}"

arr = [3, 1, 9, 7]
print(maxElement(arr))
```

### Problem 2 — Find the Minimum Element

```python
def minElement(arr):
    min = arr[0]
    for i in range(1, len(arr)):
        if min > arr[i]:
            min = arr[i]
        else:
            continue
        
    return f"The min element is {min}"


arr = [3, 1, 9, 7]
print(minElement(arr))
```

### Problem 3 — Reverse the Array (Two-Pointer Method)

```python
def reverseArray(arr):
    left = 0
    right = len(arr)-1
    
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        
        left += 1
        right -= 1
        
    return arr

arr = [1,2,3,4,5]
print("Before", arr)
print("After ", reverseArray(arr))
```

### Problem 4 — Check if Array is Sorted

```python
def checkSort(arr):
    for i in range(len(arr)):
        if arr[i] > arr[i+1]:
            return False
        else:
            return True
        

arr = [3, 1, 9, 7]
print(checkSort(arr))
```

### 6. Count frequency of each element

```python
def count_freq_sorted(arr):
    count = 1
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            count += 1
        else:
            print(arr[i-1], "->", count)
            count = 1
    print(arr[-1], "->", count)
    
    
count_freq_sorted([1,1,2,2,2,3,4,4])
```

### 8. Remove duplicates

```python
def removeDuplicate(arr):
    distinct = []
    l = 0
    for r in range(0, len(arr)):
        if arr[r] != arr[r-1]:
            distinct.append(arr[r])
            l +=1
            
    return distinct

print(removeDuplicate([1,1,2,2,2,3,4,4]))
```

### 9. Move all zeros to the end

```python
def moveZero(arr):
    slow = 0
    for fast in range(len(arr)):
        if arr[fast]:   # checks if not null or zero or false
            arr[slow], arr[fast] = arr[fast], arr[slow]
            slow += 1
            
    return arr

print(moveZero([0,1,0,3,12]))
```

### 10. Rotate array by k positions

```python

def rotateAtK(arr, k):
    
    '''Brute Force'''
    # rotate = k % len(arr)
    # for _ in range(rotate):
    #     e = arr.pop()
    #     arr.insert(0, e)
        
    # return arr

    '''Using Slicing'''
    # rotate = k%len(arr)
    # arr[:] = arr[len(arr)-rotate:]+arr[:len(arr)-rotate]
    # return arr
    
    
    '''Optimized solution'''
    n = len(arr)
    
    def reverseArr(arr, left, right):
        while left < right:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1

    
    reverseArr(arr, n-k, n-1)   # Reverse last k elements
    reverseArr(arr, 0, n-k-1)   # Reverse remaining elements
    reverseArr(arr, 0, n-1)     # Reverse whole arr
    
    return arr
    
    
print(rotateAtK([1,2,3,4,5,6,7], 4))
```

### Search in the sorted array:

```python
def getTarget(arr, target):
    left = 0
    right = len(arr)-1
    
    while left <= right:
        mid = int(left + (right-left)/2)
        
        if arr[mid] == target:
            return mid
        
        if arr[left] <= arr[mid]:
            if arr[left] <= target <= arr[mid]:
                right = mid - 1
                
            else:
                left = mid + 1
                
        else:
            if arr[mid] <= target <= arr[right]:
                left = mid + 1
            else:
                right = mid - 1
                

print(getTarget([5,6,7,1,2,3,4], 1))
```

### 11 - Count the Number of Even Elements in an Array

```python
def evenCount(arr):
    count = 0
    for i in range(len(arr)):
        if arr[i]%2 == 0:
            count += 1
            
    return count

print(evenCount([1,5,4,8,10,22,6,47]))
```

### 13 - Count how many numbers are greater than a given target

```python
def greaterTarget(arr, target):
    count = 0
    for i in range(len(arr)):
        if arr[i] > target:
            count += 1
            
    return count

print(greaterTarget([1,5,4,8,10,22,6,47], 10))
```

### 14 - Sum of All Positive Numbers in an Array

```python
def  sumPositive(arr):
    total_sum = 0
    for i in range(len(arr)):
        if arr[i] > 0:
            total_sum+=arr[i]
            
    return total_sum

print(sumPositive([-200, +500, -100, +300, -50, +100]))
```

### 15 - Count how many elements are divisible by 3

```python
def divisible3(arr):
    count = 0
    for i, num in enumerate(arr):
        # return num
        if num%3 == 0:
            count += 1
    return count

print(divisible3([3, 5, 9, 10, 12, 14]))
```

### 16 - Find the sum of all elements in an array

```python
def sumElement(arr):
    total_sum = 0
    for i, num in enumerate(arr):
        total_sum += num
        
    return total_sum

print(sumElement([200, 150, 400, 50]))
```

### 17 - Count how many numbers are within a given range [L, R]

```python
def inRange(arr, l, r):
    count = 0
    for i, num in enumerate(arr):
        if num>=l and num<=r:
            count += 1
            
    return count

print(inRange([3, 5, 9, 10, 12, 14], 2,6))
```
