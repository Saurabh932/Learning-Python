```python
class queue:
    def __init__(self):
        self.values = []
        
    def enqueue(self, value):
        self.values.append(value)
        print(self.values)
        
    def dequeue(self):
        front = self.values[0]
        self.values = self.values[1:]
        print(self.values)
        return front

    
q = queue()
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
q.enqueue(50)
q.enqueue(60)

q.dequeue()
q.dequeue()
q.dequeue()
q.dequeue()
```

# Queue Using deque

##### Create a Queue

```python
from collections import deque

q = deque()
print(q)
```

##### Enqueue (Insert at back)

```python
q.append(10)
q.append(20)
q.append(30)
q.append(40)
print(q)
```

##### Dequeue (Remove from front)

```python
print(q.pop())  # Removes from right or end
print(q)
print(q.popleft())   # Removes from leht or start
print(q)
```

##### Peek front element

```python
print(q[0])
```

##### Check if queue is empty

```python
print(len(q) == 0)
```
