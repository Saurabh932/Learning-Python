#### Reversing strings

```python
def revStr(s):
    ''' Method 1 '''
    # return s[::-1]
    
    ''' Method 2 '''
    rev = ""
    for i in range(len(s)-1, -1, -1):
        rev += s[i]
        
    return rev

s = input("Enter the string: ")
print(revStr(s))
```

#### Counting characters & frequency


```python
def freq(s):
    s_dict = {}
    
    for i in s:
        if i in s_dict:
            s_dict[i] += 1
        else:
            s_dict[i] = 1
            
    return s_dict, max(s_dict, key=s_dict.get)

print(freq("Saurabh"))
```

#### Palindrome check

```python
def palindrome(s):
    ''' Method 1 '''
    # return True if s[::-1]==s else False
    
    ''' Method 2 '''
    # rev_s = ""
    # for i in range(len(s)-1,-1,-1):
    #     rev_s += s[i]
        
    # return True if rev_s==s else False
        
    ''' Method 3 '''
    l = 0
    r = len(s)-1
    while l < r:
        if not s[l].isalnum():
            l += 1
        elif not s[r].isalnum():
            r -= 1
        if s[l] == s[r]:
            l += 1
            r -= 1
        else:
            return False
        
    return True
    
print(palindrome("race car"))
```

#### Anagrams

```python
def anagrams(s1, s2):
    return True if sorted(s1.lower())==sorted(s2.lower()) else False

print(anagrams("garden", "danger"))
```

#### Longest substring problems

```python
def substring(s):
    present = set()
    left = 0
    max_len = 0
    
    for right in range(len(s)):
        while s[right] in present:
            present.remove(s[left])
            left += 1
            
        present.add(s[right])
        max_len = max(max_len, right-left+1)
        
    return max_len
            
substring("Saurabh")
```
