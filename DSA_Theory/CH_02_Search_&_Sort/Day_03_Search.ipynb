## <b>Sorting

#### <b>1. Bubble Sort

âœ” Intuition

Imagine students standing in a line.
You repeatedly compare adjacent students and swap them if they are in the wrong order.

ðŸ“Œ Real-Life Example

Sorting a handful of playing cards in your hand by repeatedly moving the wrong card up.

âœ” Diagram

Array: [4, 2, 5, 1]

Pass 1:

4 > 2 â†’ swap â†’ [2, 4, 5, 1]
4 < 5 â†’ ok
5 > 1 â†’ swap â†’ [2, 4, 1, 5]


Pass 2:

2 < 4
4 > 1 â†’ swap â†’ [2, 1, 4, 5]


Pass 3:

2 > 1 â†’ swap â†’ [1, 2, 4, 5]

```python
def bubbleSort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    
    return arr
            
print(bubbleSort([4, 2, 5, 1]))
```

#### <b>2. Selection Sort

âœ” Intuition

You select the smallest element and put it at the beginning.
Then repeat for the next position.

ðŸ“Œ Real-life Example

Finding the cheapest product in a store, marking it as first.
Then finding the next cheapest, and so on.

âœ” Diagram

Array: [3, 1, 4, 2]

Step 1: find min â†’ 1 â†’ swap with index 0
â†’ [1, 3, 4, 2]

Step 2: find min in remaining â†’ 2 â†’ swap with index 1
â†’ [1, 2, 4, 3]

Step 3: find min in remaining â†’ 3 â†’ swap with index 2
â†’ [1, 2, 3, 4]


```python
def selectionSort(arr):
    for i in range(len(arr)-1):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        
        # Swap
        arr[min_idx], arr[i] = arr[i], arr[min_idx]
        
    return arr

print(selectionSort([4, 2, 5, 1]))
```

#### <b>3. Insertion Sort
âœ” Intuition

You build the sorted array one element at a time.

ðŸ“Œ Real-life Example

Sorting cards in your hand while playing rummy.
You pick a card and insert it in the correct position among already sorted cards.

âœ” Diagram

Array: [4, 3, 2, 1]

Steps:

Insert 3 into [4] â†’ [3,4]
Insert 2 into [3,4] â†’ [2,3,4]
Insert 1 â†’ [1,2,3,4]>

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        curr = arr[i]
        prev = i-1
        
        # Finding out the current position to insert
        while prev >= 0 and arr[prev] > curr:
            arr[prev+1] = arr[prev]
            prev -= 1
            
        # insertion
        arr[prev+1] = curr
        
    return arr

print(insertionSort([4, 2, 5, 1]))
```
