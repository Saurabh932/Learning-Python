### 1. What is an Array?
In Python, we normally use lists as arrays because they are dynamic and can store mixed types.

```python
# Empty array
arr = []
```

```python

# Array wuth value
arr = [1,2,3,4,5]
print(arr)
```

```python
# Array using list comprehension
arr = [i for i in range(5)]
print(arr)
```

```python
# Array with repeated value
arr = [0]*5
print(arr)
```

### 2. Using the array Module (Real fixed-type array)
Python has a built-in module array which stores elements of a single data type, like C arrays.

| Code  | Type       |
| ----- | ---------- |
| `'i'` | signed int |
| `'f'` | float      |
| `'d'` | double     |


```python
# Empty array
from array import array
arr = array('i')
print(arr)
```

```python
from array import array

arr = array("i", [1,2,3])
print(arr)
```

### 3. Using NumPy Arrays

```python
import numpy as np

np_arr = np.array([1,2,3])
print(np_arr)
```

### 4. Using * unpack operator

```python
a = [1, 2]
b = [3, 4]

print([*a, *b])
```

### 5. Using map() function

```python
arr = list(map(int, input("Enter the element").strip().split(",")))
print(arr)
```
