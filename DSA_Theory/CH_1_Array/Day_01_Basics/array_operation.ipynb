### Accessing Elements

```python
arr = [1,2,3,4]

print(arr[0])
print(arr[3])
```

### Updating Element

```python
arr[1] = 5
print(arr) 
```

### Inserting Element

```python
# Insert at the end

arr.append(7)
print(arr)
```

```python
# Insert at the specific posiution    

arr.insert(2, 25)
print(arr)
```

```python
# Insert multiple elements

arr.extend([8,9,10])
print(arr)
```

### Deleting Elements

```python
print(arr.pop())
```

```python
print(arr.pop(3))
print(arr)
```

### Delete by value

```python
print(arr)
print(arr.remove(25))
print(arr)
```

### Searching

```python
# Linear Search -> O(n)

def linear_search(arr, target):
    for ele in range(len(arr)):
        if arr[ele] == target:
            return f"Found {target} at {ele} index."
        else:
            continue
    return f"{target} not present in array"
        
print(linear_search(arr=[1,2,3,4], target=5))
```
