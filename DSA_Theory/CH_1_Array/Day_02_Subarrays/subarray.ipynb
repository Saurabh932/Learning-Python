#### 1. What is a Subarray?

A subarray is:
 - A continuous part of an array
 - Elements must be from contiguous positions

Example: n = 5 ‚Üí 15 subarrays.

```python
def subArray(arr):
    for i in range(len(arr)):
        for j in range(i+1, len(arr)+1):
            print(arr[i:j], end=" ")
        print()
        
subArray([1,2,3,4,5])
```

#### Sum of Subarrays (Kadane‚Äôs algorithm)

```python
def sum_subarr(arr):
    for i in range(len(arr)):
        total = 0
        for j in range(i, len(arr)):
            total += arr[j]
            print(total, end=" ")
        print()
            
sum_subarr([1,2,3,4])
```

#### Generate the Maximum Subarray Sum

```python
def maxSumSubarr(arr):
    maxSum = float('-inf')
    total = 0
    for i in range(len(arr)):
        total += arr[i]
        maxSum = max(total, maxSum)
            
    return f"Max Sum of Subarray {maxSum}"

print(maxSumSubarr([1,2,3,4]))
```
