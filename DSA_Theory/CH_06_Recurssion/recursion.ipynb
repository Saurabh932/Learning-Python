#### What is Recursion?

Recursion is a programming technique where a function calls itself to solve smaller instances of the same problem.<br></br>
<b>A recursive function must have:</b>

1ï¸âƒ£ Base Case
â€“ The condition where the function stops calling itself.

2ï¸âƒ£ Recursive Case
â€“ The part where the function calls itself with a smaller/simpler input.

ðŸ“Œ Without a base case â†’ infinite recursion â†’ program crash (RecursionError).

```python
def countdown(n):
    if n == 0:
        print("Go!!")
        return
    print(n)
    countdown(n-1)
    
countdown(10)
```

#### Print number from n to 1

```python
def desc(num):
    if num == 1:
        print(1)
        return
    
    print(num)
    
    desc(num-1)
    
desc(10)
```

#### Print number from 1 to n

```python
def asc(num):
    if num == 1:
        print(1)
        return
    
    asc(num-1)
    print(num)
    
asc(10)
```

#### Factorial

```python
def factorial(num):
    if num == 0:
        return 1
    
    return num * factorial(num-1)
    
factorial(5)
```

#### Sum of all nautral number

```python
def naturalSum(num):
    if num == 1:
        return 1
    
    return num + naturalSum(num-1)

naturalSum(5)
```

##### Fibonacci Number

```python
def fibonacci(num):
    if num == 0:
        return 0
    if num == 1:
        return 1
    
    return fibonacci(num-1) + fibonacci(num-2)

fibonacci(10)
```

```python
def sortedArr(arr, i):
    if i == len(arr)-1:
        return True
    
    if arr[i] > arr[i+1]:
        return False
    
    return sortedArr(arr, i+1)

print(sortedArr([1,2,3,4,5], 0))
```

##### First Occurance

```python
def firstOccur(arr, i, k):
    if i == len(arr):
        return "Not found"
    
    if arr[i] == k:
        return i
    
    return firstOccur(arr, i+1, k)
    
print(firstOccur([2], 0, 3))
```

##### Last Occurance

```python
def lastOccur(arr, i, k):
    if i == len(arr):
        return -1
    
    isFound = lastOccur(arr, i+1, k)
    
    if isFound != -1:
        return isFound
    
    if arr[i] == k:
        return i
    
    return isFound
    
print(lastOccur([-5, -1, -3, -1], 0, -1))
```

#### Print x to the power n

```python
def xPowN(x, n):
    if n == 1:
        return x
    
    return x * xPowN(x, n-1)

print(xPowN(2, 4))
```

### Optimized

```python
def optimize(x, n):
    if n == 0:
        return 1
    
    halfPowerSq = optimize(x, n//2) * optimize(x, n//2)
    
    if n%2 != 0:
        halfPowerSq = x * halfPowerSq
        
    return halfPowerSq

print(optimize(2, 10))
```

#### Tiling Problem

```python
def tiling(x, n):
    if n == 0 or n == 1:
        return 1
    
    # verical choice
    vertical = tiling(x, n-1)
    
    # Horizontal choice
    horizontal = tiling(x, n-2)
    
    return vertical + horizontal

print(tiling(2, 4))
```

```python
def strDuplicate(s, i):
    if i == len(s):
        return -1
    x = ""
    
    if s[i] not in x:
        x.join(s[i])
        
    return x
        
print(strDuplicate("ssggeetgwvsst", 0))
```
