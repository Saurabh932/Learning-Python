#### What is Recursion?

Recursion is a programming technique where a function calls itself to solve smaller instances of the same problem.<br></br>
<b>A recursive function must have:</b>

1Ô∏è‚É£ Base Case
‚Äì The condition where the function stops calling itself.

2Ô∏è‚É£ Recursive Case
‚Äì The part where the function calls itself with a smaller/simpler input.

üìå Without a base case ‚Üí infinite recursion ‚Üí program crash (RecursionError).

```python
def countdown(n):
    if n == 0:
        print("Go!!")
        return
    print(n)
    countdown(n-1)
    
countdown(10)
```

#### Print number from n to 1

```python
def desc(num):
    if num == 1:
        print(1)
        return
    
    print(num)
    
    desc(num-1)
    
desc(10)
```

#### Print number from 1 to n

```python
def asc(num):
    if num == 1:
        print(1)
        return
    
    asc(num-1)
    print(num)
    
asc(10)
```

#### Factorial

```python
def factorial(num):
    if num == 0:
        return 1
    
    return num * factorial(num-1)
    
factorial(5)
```

#### Sum of all nautral number

```python
def naturalSum(num):
    if num == 1:
        return 1
    
    return num + naturalSum(num-1)

naturalSum(5)
```

##### Fibonacci Number

```python
def fibonacci(num):
    if num == 0:
        return 0
    if num == 1:
        return 1
    
    return fibonacci(num-1) + fibonacci(num-2)

fibonacci(10)
```

```python
def sortedArr(arr, i):
    if i == len(arr)-1:
        return True
    
    if arr[i] > arr[i+1]:
        return False
    
    return sortedArr(arr, i+1)

print(sortedArr([1,2,3,4,5], 0))
```

##### First Occurance

```python
def firstOccur(arr, i, k):
    if i == len(arr):
        return "Not found"
    
    if arr[i] == k:
        return i
    
    return firstOccur(arr, i+1, k)
    
print(firstOccur([2], 0, 3))
```

##### Last Occurance

```python
def lastOccur(arr, i, k):
    if i == len(arr):
        return -1
    
    isFound = lastOccur(arr, i+1, k)
    
    if isFound != -1:
        return isFound
    
    if arr[i] == k:
        return i
    
    return isFound
    
print(lastOccur([-5, -1, -3, -1], 0, -1))
```

#### Print x to the power n

```python
def xPowN(x, n):
    if n == 1:
        return x
    
    return x * xPowN(x, n-1)

print(xPowN(2, 4))
```

### Optimized

```python
def optimize(x, n):
    if n == 0:
        return 1
    
    halfPowerSq = optimize(x, n//2) * optimize(x, n//2)
    
    if n%2 != 0:
        halfPowerSq = x * halfPowerSq
        
    return halfPowerSq

print(optimize(2, 10))
```

#### Tiling Problem

```python
def tiling(x, n):
    if n == 0 or n == 1:
        return 1
    
    # verical choice
    vertical = tiling(x, n-1)
    
    # Horizontal choice
    horizontal = tiling(x, n-2)
    
    return vertical + horizontal

print(tiling(2, 4))
```

```python
def strDuplicate(s, i, x=""):
    if i == len(s):
        return x
    
    curr = s[i]
        
    if curr not in x:
        return strDuplicate(s, i, x+curr)
    
    else:
        return strDuplicate(s, i+1, x)
        
print(strDuplicate("ssggeetgwvsst", 0))
```

### Find sum of digits of a number using recursion

```python
def digitSum(num):
    if num == 0:
        return 0
    
    return (num%10) + digitSum(num//10)

print(digitSum(12345))
```

#### Check if a string is a palindrome

```python
def isPalindrome(s, left, right):
    # Base case: if pointers cross or meet ‚Üí it's a palindrome
    if left >= right:
        return True
    
    # If mismatch found ‚Üí not a palindrome
    if s[left] != s[right]:
        return False
    
    # Recursive step: move inward
    return isPalindrome(s, left + 1, right - 1)


word = "madam"
print(isPalindrome(word, 0, len(word)-1))
```

### Reverse a string using recursion

```python
def revStr(s, i):
    if i == -1:
        return ""

    return s[i] + revStr(s, i-1)

print(revStr("abcd", len("abcd")-1))
```

### Sum of array elements using recursion

```python
def sumArr(arr, i):
    if i>len(arr)-1:
        return 0
    
    return arr[i] + sumArr(arr, i+1)

print(sumArr([1,2,3,4,5], 0))
```

### Count how many times a target appears in an array

```python
def countTar(arr, i, target):
    if i == len(arr)-1:
        return 0
    
    count = 0
    if arr[i] == target:
        count += 1

    
    return count + countTar(arr, i+1, target)

print(countTar([1,3,3,4,3,5], 0, 4))
```

### Find the maximum element in an array

Call ‚Üì                     | Return ‚Üë
-------------------------- | -----------------------------
maxEle(0)   arr[0] = 1     | ‚Üê  max(1, 89)  = 89
maxEle(1)   arr[1] = 3     | ‚Üê  max(3, 89)  = 89
maxEle(2)   arr[2] = 89    | ‚Üê  max(89, 5) = 89
maxEle(3)   arr[3] = 3     | ‚Üê  max(3, 5)  = 5
maxEle(4)   arr[4] = 4     | ‚Üê  max(4, 5)  = 5
maxEle(5)   arr[5] = 3     | ‚Üê  max(3, 5)  = 5
maxEle(6)   arr[6] = 5     | ‚Üí  base case returns 5


```python
def maxEle(arr, i):
    if i == len(arr)-1:
        return arr[i]
    
    last= maxEle(arr, i+1)
    return max(arr[i], last)

print(maxEle([1,3, 89,3,4,3,5], 0))
```

```python
def minEle(arr, i):
    if i == len(arr)-1:
        return arr[i]
    
    last= minEle(arr, i+1)
    return min(arr[i], last)

print(minEle([1,3,-1, 89,3,4,3,5], 0))
```
