### Count frequency of each element

```python
def freqCount(arr):
    count = {}
 
    for num in arr:
        if num in count:
            count[num] += 1
            
        else:
            count[num] = 1
         
    return count

print(freqCount([1, 2, 2, 3, 3, 3,2,2]))
```

### Find the first non-repeating element

```python
def nonRepeat(arr):
    count = {}
    
    for num in arr:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1
            
    for num in count:
        if count[num] == 1:
            return num
        
    return -1

arr = [2, 3, 4, 2, 3, 5]
print(nonRepeat(arr))
```

### Check if two arrays contain the same elements (regardless of order)


```python
def sameArr(arr1, arr2):
    if len(arr1) != len(arr2):
        return False
    
    freq = {}
    
    for key in arr1:
        freq[key] = freq.get(key, 0) + 1
            
    for key in arr2:
        if key not in freq or freq[key] == 0:
            return False
        freq[key] -= 1
        
    return all(value == 0 for value in freq.values())
        
print(sameArr([1,2,2,3, 7], [1,2,2,3, 4]))
```

```python
def maxCount(arr):
    count = {}
 
    for num in arr:
        if num in count:
            count[num] += 1
            
        else:
            count[num] = 1
         
    return max(count, key=count.get)

print(maxCount([1, 2, 2, 3, 3, 3,2,2]))
```

### Find all elements that appear more than once

```python
def morethan1(arr):
    count = {}
    duplicate = []
    for key in arr:
        count[key] = count.get(key, 0) + 1
            
    for key in count:
        if count[key] > 1:
            duplicate.append(key)
    return duplicate

print(morethan1([1,2,3,2,4,3,3]))
```

### Count how many elements appear exactly once

```python
def appearOnce(arr):
    count = {}
    once = []
    countOnce = 0
    
    for key in arr:
        count[key] = count.get(key, 0) + 1
        
    for key in count:
        if count[key] == 1:
            countOnce += 1
            once.append(key)
            
    return countOnce, once

print(appearOnce([1,2,2,3,4,4,5]))
```
