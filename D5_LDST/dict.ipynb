```python
my_map = {"a": 1, "b": 2, "c": 3}

for key in my_map:    # for key in my_map.key():
    print(key)
```

```python
my_map = {"a": 1, "b": 2, "c": 3}

for value in my_map.values():
    print(value)
```

```python
my_map = {"a": 1, "b": 2, "c": 3}

for key, value in my_map.items():
    print(key)
    print(value)
```

```python
my_map = {"a": 1, "b": 2, "c": 3}

for i, (key, value) in enumerate(my_map.items()):
    print(i, key, value)
```

```python
my_map = {"a": 1, "b": 2, "c": 3}

for key in sorted(my_map):
    print(key, my_map[key])
```

#### Iterating while modifying a dict

```python
my_map = {"a": 1, "b": 2, "c": 3}

for key, value in my_map.copy().items():
    if value % 2 == 0:
        my_map[key] = 0
        
print(my_map) 
```

#### Iterating through nested dictionaries

```python
people = {
    "alice": {"age": 30, "city": "London"},
    "bob": {"age": 25, "city": "Paris"}
}


for person, detail in people.items():
    print("Person: ", person)
    for key, value in detail.items():
        print(" ", key, value)
```

```python
arr = [3, 5, 2, 4]

for i, num in enumerate(arr):
    print(i, num)
```
